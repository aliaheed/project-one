```python
# load libraries
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly
import plotly.plotly as py
import plotly.graph_objs as go
```

```python
# Load data
fp = "../Resources/Fire_Data_Over_1.csv"

map_df = pd.read_csv(fp)

map_df.head()
```

```python
# Remove Puerto Rico since it won't show up on our map

new = map_df.loc[(map_df.loc[:,'STATE']!='PR')]
new.head()
```

```python
# Sort by year ascending 

df = new.sort_values(["FIRE_YEAR", "STATE"])
df.dropna(how="any")
df = df.reset_index(drop=True)
df.head()
```

```python
# Simplify to two columns and count number of wildfires in each year

group_df = df[['FIRE_YEAR','STATE']]
state_totals = list(df.groupby("STATE")["FIRE_YEAR"].count())
year_totals = list(group_df["FIRE_YEAR"].value_counts())
```

```python
# Make a unique data frame to pull unique years and states

unique_df = group_df.drop_duplicates().reset_index(drop=True)
unique_states = group_df.loc[group_df["FIRE_YEAR"] == 2015].drop_duplicates().reset_index(drop=True)
```

```python
states = unique_states["STATE"]
years = list(range(1992,2016))

# Make a df of totals for the years

years_df = pd.DataFrame({
    "Year": years,
    "Totals": year_totals
})
years_df.head()
```

```python
# Make a df of totals for the states

states_df = pd.DataFrame({ 
    "State": states, 
    "Totals": state_totals
})
states_df.head()
```

```python
# Get the number of wildfires in each year in each state

df = df[["FIRE_YEAR", "STATE"]]
totals = list(df.groupby(["FIRE_YEAR", "STATE"])["STATE"].count())
unique_df["Totals"] = totals
unique_df.head()
```

```python
years = list(range(1992,2016))

for row in unique_df, year in years:
     if unique_df["FIRE_YEAR"] == year:
            states_df.loc[unique_df.index[row], f'{year}'] = unique_df["Totals"].loc[unique_df["FIRE_YEAR"] == year]

states_df.head()
```

```python
# Find the totals for each year and stick them in the new columns

states_df["1992"] = unique_df["Totals"].loc[unique_df["FIRE_YEAR"] == 1992]
states_df["1993"] = unique_df["Totals"].loc[unique_df["FIRE_YEAR"] == 1993]
states_df["1994"] = unique_df["Totals"].loc[unique_df["FIRE_YEAR"] == 1994]
states_df["1995"] = unique_df["Totals"].loc[unique_df["FIRE_YEAR"] == 1995]
states_df["1996"] = unique_df["Totals"].loc[unique_df["FIRE_YEAR"] == 1996]
states_df["1997"] = unique_df["Totals"].loc[unique_df["FIRE_YEAR"] == 1997]
states_df["1998"] = unique_df["Totals"].loc[unique_df["FIRE_YEAR"] == 1998]
states_df["1999"] = unique_df["Totals"].loc[unique_df["FIRE_YEAR"] == 1999]
states_df["2000"] = unique_df["Totals"].loc[unique_df["FIRE_YEAR"] == 2000]
states_df["2001"] = unique_df["Totals"].loc[unique_df["FIRE_YEAR"] == 2001]
states_df["2002"] = unique_df["Totals"].loc[unique_df["FIRE_YEAR"] == 2002]
states_df["2003"] = unique_df["Totals"].loc[unique_df["FIRE_YEAR"] == 2003]
states_df["2004"] = unique_df["Totals"].loc[unique_df["FIRE_YEAR"] == 2004]
states_df["2005"] = unique_df["Totals"].loc[unique_df["FIRE_YEAR"] == 2005]
states_df["2006"] = unique_df["Totals"].loc[unique_df["FIRE_YEAR"] == 2006]
states_df["2007"] = unique_df["Totals"].loc[unique_df["FIRE_YEAR"] == 2007]
states_df["2008"] = unique_df["Totals"].loc[unique_df["FIRE_YEAR"] == 2008]
states_df["2009"] = unique_df["Totals"].loc[unique_df["FIRE_YEAR"] == 2009]
states_df["2010"] = unique_df["Totals"].loc[unique_df["FIRE_YEAR"] == 2010]
states_df["2011"] = unique_df["Totals"].loc[unique_df["FIRE_YEAR"] == 2011]
states_df["2012"] = unique_df["Totals"].loc[unique_df["FIRE_YEAR"] == 2012]
states_df["2013"] = unique_df["Totals"].loc[unique_df["FIRE_YEAR"] == 2013]
states_df["2014"] = unique_df["Totals"].loc[unique_df["FIRE_YEAR"] == 2014]
states_df["2015"] = unique_df["Totals"].loc[unique_df["FIRE_YEAR"] == 2015]


```

```python
# Now to make cloropleth maps 

for col in states_df.columns:
    states_df[col] = states_df[col].astype(str)

scl = [[0.0, 'rgb(255,215,0)'],
       [0.2, 'rgb(255,165,0)'],
       [0.4, 'rgb(255,140,0)'],
       [0.6, 'rgb(255,115,0)'],
       [0.8, 'rgb(255,90,0)'],
       [1.0, 'rgb(255,65,0)']]

df['text'] = states_df['State']

data = [ dict(
        type='choropleth',
        colorscale = scl,
        autocolorscale = False,
        locations = states_df['State'],
        z = states_df['Totals'].astype(float),
        locationmode = 'USA-states',
        text = df['text'],
        marker = dict(
            line = dict (
                color = 'rgb(255,255,255)',
                width = 2
            ) ),
        colorbar = dict(
            title = "Number of Wildfires")
        ) ]

layout = dict(
        title = 'USA Wildfires 1992 -2015 </br></br>Source: Kaggle',
        geo = dict(
            scope='usa',
            projection=dict( type='albers usa' ),
            showlakes = True,
            lakecolor = 'rgb(255, 255, 255)'),
             )
    
fig = dict( data=data, layout=layout )
py.iplot( fig, filename='d3-cloropleth-map' )
```

```python
#Bonus: make a for loop to save each year's map as png

import plotly.io as pio

static_image_bytes = pio.to_image(fig, format='png')

from IPython.display import Image

Image(static_image_bytes)

pio.write_image(fig, file='uswildfire_all_years.png', format='png')
```
